```python
cols=['Loan_ID','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Total_Applicant_Income']

df = df.drop(columns=cols,axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
colss=['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
for i in colss:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

```python
df.dtypes
```

```python
X=df.drop('Loan_Status',axis=1)
Y=df['Loan_Status']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.25,random_state = 42)
```

```python
model1=LogisticRegression()
model1.fit(X_train,Y_train)
Y_predict_model1=model1.predict(X_test)
accuracy_model1=accuracy_score(Y_test,Y_predict_model1)
print(accuracy_model1*100)
```

```python
score=cross_val_score(model1,X,Y,cv=6)
print(score*100)
print(np.mean(score)*100)
```

```python
model2=RandomForestClassifier()
model2.fit(X_train,Y_train)
Y_predict_model2=model2.predict(X_test)
accuracy_model2=accuracy_score(Y_test,Y_predict_model2)
print(accuracy_model2*100)
```

```python
score=cross_val_score(model2,X,Y,cv=6)
print(score*100)
print(np.mean(score)*100)
```

```python
model3=DecisionTreeClassifier()
model3.fit(X_train,Y_train)
Y_predict_model3=model3.predict(X_test)
accuracy_model3=accuracy_score(Y_test,Y_predict_model3)
print(accuracy_model3*100)
```

```python
score=cross_val_score(model3,X,Y,cv=5)
print(score*100)
print(np.mean(score)*100)
```

```python
model4=KNeighborsClassifier()
model4.fit(X_train,Y_train)
Y_predict_model4=model4.predict(X_test)
accuracy_model4=accuracy_score(Y_test,Y_predict_model4)
print(accuracy_model4*100)
```

```python
score=cross_val_score(model4,X,Y,cv=5)
print(score*100)
print(np.mean(score)*100)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
def generate_classification_report(model_name,Y_test,Y_predict):
    report=classification_report(Y_test,Y_predict)
    print(f"classification report for {model_name} is : \n{report}\n")

generate_classification_report(model1,Y_test,Y_predict_model1)
generate_classification_report(model2,Y_test,Y_predict_model2)
generate_classification_report(model3,Y_test,Y_predict_model3)
generate_classification_report(model4,Y_test,Y_predict_model4)
```

```python
from imblearn.over_sampling import RandomOverSampler ##to handle outliers(1s>0s)

```

```python
oversample=RandomOverSampler(random_state=42)
X_resampled,Y_resampled=oversample.fit_resample(X,Y)
df_resampled=pd.concat([pd.DataFrame(X_resampled,columns=X.columns),pd.Series(Y_resampled,name="Loan_status")],axis=1)
```

```python
X_resampled
```

```python
Y_resampled
```

```python
Y_resampled.value_counts()
```

```python
X_resampled_train, X_resampled_test, Y_resampled_train, Y_resampled_test = train_test_split(X_resampled,Y_resampled,test_size = 0.25,random_state=42)
```

```python
model1 = LogisticRegression()
model1.fit(X_resampled_train,Y_resampled_train)
y_pred_model1 = model1.predict(X_resampled_test)
accuracy = accuracy_score(Y_resampled_test,y_pred_model1)
accuracy*100
```

```python

model2 = DecisionTreeClassifier()
model2.fit(X_resampled_train,Y_resampled_train)
y_pred_model2 = model2.predict(X_resampled_test)
accuracy = accuracy_score(Y_resampled_test,y_pred_model2)
print("Accuracy score of Decision Tree: ", accuracy*100)
```

```python
model3 = RandomForestClassifier()
model3.fit(X_resampled_train,Y_resampled_train)
y_pred_model3 = model3.predict(X_resampled_test)
accuracy = accuracy_score(Y_resampled_test,y_pred_model3)
print("Accuracy score of Random Forest: ", accuracy*100)
```

```python
model4 = KNeighborsClassifier(n_neighbors=3)
model4.fit(X_resampled_train,Y_resampled_train)
y_pred_model4 = model4.predict(X_resampled_test)
accuracy = accuracy_score(Y_resampled_test,y_pred_model4)
print("Accuracy score of KNeighbors: ", accuracy*100)
```
